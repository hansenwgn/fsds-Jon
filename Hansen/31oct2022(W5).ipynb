```python
myData = {
    'id'         : [0, 1, 2, 3, 4, 5],
    'Name'       : ['Greater London', 'Greater Manchester', 'Birmingham','Edinburgh','Inverness','Lerwick'],
    'Rank'       : [1, 2, 3, 4, 5, 6],
    'Longitude'  : [-0.128, -2.245, -1.903, -3.189, -4.223, -1.145],
    'Latitude'   : [51.507, 53.479, 52.480, 55.953, 57.478, 60.155],
    'Population' : [9787426, 2705000, 1141816, 901455, 70000, 6958],
}

df = pd.DataFrame(myData)
```

```python
print(df)
print()
print(df.describe())
print()
print(df.info())
```

```python
df.set_index('id') #choose the unique, distinct value
```

```python
df.set_index('id', inplace=True) #why we need inplace? we don't want to modify the original data unless we want to
```
